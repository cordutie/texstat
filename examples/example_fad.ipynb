```python
# Add the parent directory to the Python path
import sys
import os
parent_dir = os.path.abspath('..')
if parent_dir not in sys.path:
    sys.path.append(parent_dir)

# Import texstat packages
from texstat.fad import FAD_wrapper
```

```python
# Texstat model parameters
sr            = 44100
frame_size    = sr*2 # 2-second frames
hop_size      = sr   # 1-second hop (50% overlap)
class_1 = "../classes/g_Bees/"
class_2 = "../classes/o_Bees/"

FAD_texstat = FAD_wrapper(frame_size = frame_size, hop_size = hop_size, sampling_rate = sr, device='cpu')

# Compute FAD score using the full set of statistics
fad_score = FAD_texstat.score(class_1, class_2, cache_dir='embeddings_full', model="full")
print(f"FAD score: {fad_score:.4f}")

# Compute FAD score using the subset of statistics
fad_score_subset = FAD_texstat.score(class_1, class_2, cache_dir='embeddings_sub', model="subset")
print(f"FAD score (subset): {fad_score_subset:.4f}")
```
